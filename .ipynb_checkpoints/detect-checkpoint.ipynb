```python
DATA_PATH = "./test-data/data1"
ANGLE_ACCURACY = 100
```

```python
import json
import sys
import numpy as np
from math import *
from matplotlib import pyplot as plt
```

```python
print("Processing data in {}".format(DATA_PATH))
```

```python
with open(DATA_PATH + "/info.json") as f:
    raw_data = json.load(f)
    angle = raw_data["angle"]
    n_frames = raw_data["n-frames"]
    frame_width, data_width, height = raw_data["frame-width"], raw_data["data-width"], raw_data["height"]
```

```python
print("Info: ", raw_data)
```

```python
def getAngleStrength(img):
    angle_strength = np.zeros(ANGLE_ACCURACY)
    for x in range(frame_width):
        for y in range(height):
            theta = atan2(y-height//2, x-frame_width//2)

            theta %= pi
            theta = pi - theta - 0.000001
            angle_strength[int(round( ANGLE_ACCURACY * ((theta/pi)) ))%ANGLE_ACCURACY] += img[x][y]
    
    return angle_strength
```

```python
angle_strengths = np.ndarray(n_frames*ANGLE_ACCURACY)
for i in range(n_frames):
    print("Processing frame {}".format(i))
    with open(DATA_PATH + "/frame{}.json".format(i)) as f:
        frame = np.array(json.load(f)["frame"])
        strengths = getAngleStrength(frame)
        
        for j in range(len(strengths)):
            angle_strengths[i * ANGLE_ACCURACY + j] = strengths[j]
        
        
```

```python
angle_strengths = angle_strengths.reshape((n_frames, ANGLE_ACCURACY))
print(angle_strengths)

fig = plt.figure()
ax = plt.axes(projection='3d')
X, Y = np.meshgrid([i for i in range(ANGLE_ACCURACY)], [i for i in range(n_frames)])
ax.plot_surface(X, Y, angle_strengths)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

```python
m = max(angle_strength)
detected = pi * np.where(angle_strength == m)[0]/ANGLE_ACCURACY
print("Detected Angle: {}".format(detected))
```
